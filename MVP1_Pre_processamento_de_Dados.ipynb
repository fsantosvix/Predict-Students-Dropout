## 1. Contexto e Informações Gerais

### 1.1 Descrição do Problema

A evasão estudantil é um dos principais desafios enfrentados por universidades e faculdades, afetando a eficiência do ensino e até mesmo a reputação dessas instituições. Esse fenômeno, frequentemente denominado *student churn*, refere-se à saída de estudantes antes da conclusão do curso. Manter os estudantes engajados e garantir sua permanência até a conclusão do curso envolve fatores acadêmicos, sociais, econômicos e individuais que nem sempre são fáceis de identificar.

O objetivo deste estudo é antecipar o risco de evasão, utilizando dados históricos e características conhecidas dos estudantes ao longo de sua trajetória acadêmica. A partir dessa análise, espera-se contribuir para que universidades e faculdades possam identificar perfis de risco com maior precisão e, futuramente, fundamentar ações proativas voltadas à retenção estudantil.

### 1.2 Hipóteses do Problema

As hipóteses que tratarei nesse estudo que envolve apenas a etapa de Pré-Processamento dos dados são as seguintes:

1. A idade dos estudantes interfere na nota de admissão?

2. A nota obtida em qualificação anterior é indicativo da nota de admissão no curso estudado posteriormente?

3. Estudantes com necessidades educacionais especiais apresentam maior taxa de evasão?

### 1.3 Tipo de Problema

Dentro de um escopo mais abrangente, trata-se de um problema de **classificação supervisionada**. Dado um conjunto de características detalhadas no decorrer do estudo, busca-se prever se o estudante tende a abandonar o curso ou permanecer matriculado (ou concluir o curso) decorrida a duração normal da qualificação.

## 3. Declaração das Funções Usadas no Código

```python
# Cria uma função para contagem de valores zerados em determinadas colunas
def contar_zeros(df, colunas):
    for col in colunas:
        qtd_zeros = (df[col] == 0).sum()
        print(f'A quantidade de zeros para {col} é {qtd_zeros}')

# Cria uma função frequency encoding apra múltiplas colunas
def frequency_encoding(X, cols):
  X_ = X.copy()
  for col in cols:
    freqs = X_[col].value_counts(normalize=True) # calcula frequencia relativa
    X_[col] = X_[col].map(freqs).fillna(0) # substitui os códigos pela frequencia
  return X_

# Cria função que será usada no pipeline de transformação de dados
def apply_frequency_encoding(X):
  df_freq = pd.DataFrame(X, columns=col_frequency)
  return frequency_encoding(df_freq, col_frequency)
```

### 6.4 Transformação dos Dados

Diante das características dos atributos, identifiquei a necessidade de aplicar as seguintes técnicas de pré-processamento e transformação:

- Normalização ou Padronização - para atributos numéricos contínuos com grande amplitude.

- One hot encoding - para variáveis categóricas convertidas em código com até 10 valores distintos.
- Frequency encoding - para variáveis categóricas convertidas em código mas com mais de 10 valores distintos.

O tratamento das variáveis categóricas se faz necessário mesmo que elas já tenham sido convertidas para código para evitar que o modelo entenda que há uma relação matemática entre os valores. Por exemplo, o curso 9254 não deve ser intepretado como "maior" que o curso 0171.

Há ainda um conjunto de colunas binárias que não sofrerão tratamento adicional.

Para permitir a aplicação de cada uma das técnicas listadas acima nos respectivos atributos, devemos separar as colunas de acordo com o tipo de processamento que sofrerão.

```python
# Definição de cada grupo de atributos a serem tratados, conforme o tipo de processamento a ser aplicado
col_normalizacao = ['Application order',
                    'Previous qualification (grade)',
                    'Admission grade',
                    'Age at enrollment',
                    'Unemployment rate',
                    'Inflation rate',
                    'GDP'
                    ]

col_pca = ['Curricular units 1st sem (enrolled)',
       'Curricular units 1st sem (evaluations)',
       'Curricular units 1st sem (approved)',
       'Curricular units 1st sem (grade)',
       'Curricular units 2nd sem (enrolled)',
       'Curricular units 2nd sem (evaluations)',
       'Curricular units 2nd sem (approved)',
       'Curricular units 2nd sem (grade)'
       ]

col_hot = ['Marital status']

col_frequency = ['Application mode',
                 'Course',
                 'Previous qualification',
                 'Nacionality',
                 "Mother's qualification",
                 "Father's qualification",
                 "Mother's occupation",
                 "Father's occupation",
                 ]

col_binarias = ['Daytime/evening attendance\t',
                'Displaced',
                'Educational special needs',
                'Debtor',
                'Tuition fees up to date',
                'Gender',
                'Scholarship holder',
                'International'
                ]
```

#### Paronização ou Normalização

Vamos analisar se os atributos do primeiro conjunto serão tratados via padronização ou normalização.

Para realizar essa análise, vamos aplicar as duas técnicas separadamente e comparar como elas afetam a distribuição dos valores.

```python
# Cria uma cópia dos atributos que passarão por normalização e padronização
X_train_sub = X_train[col_normalizacao].copy()

# Aplica Padronização
scaler_pad = StandardScaler()
X_pad = pd.DataFrame(scaler_pad.fit_transform(X_train_sub), columns=col_normalizacao)

# Aplica Normalização
scaler_norm = MinMaxScaler()
X_norm = pd.DataFrame(scaler_norm.fit_transform(X_train_sub), columns=col_normalizacao)
```

Para uma comparação visual, vamos plotar as três distribuições (original, padronizada e normalizada) em histogramas lado a lado.

```python
# Configura a grade de histogramas
fig, axes = plt.subplots(nrows=7, ncols=3, figsize=(18, 25))
fig.suptitle('Comparação de Histogramas: Original x Padronizada x Normalizada', fontsize=18, y=1.02)

# Gera os histogramas
for i, col in enumerate(col_normalizacao):
    # Original
    sns.histplot(X_train_sub[col], bins=10, kde=True, ax=axes[i, 0], color='steelblue')
    axes[i, 0].set_title(f'{col} - Original')

    # StandardScaler
    sns.histplot(X_pad[col], bins=10, kde=True, ax=axes[i, 1], color='darkorange')
    axes[i, 1].set_title(f'{col} - StandardScaler')

    # MinMaxScaler
    sns.histplot(X_norm[col], bins=10, kde=True, ax=axes[i, 2], color='forestgreen')
    axes[i, 2].set_title(f'{col} - MinMaxScaler')

# Ajusta layout
plt.tight_layout()
plt.subplots_adjust(top=0.98)
plt.show()
```

Podemos perceber que há pouca variação na forma de distribuição dos valores, independente da técnica aplicada. No entanto, a NORMALIZAÇÃO apresenta uma variação maior em certos casos.

Portanto, utilizaremos a PADRONIZAÇÃO como técnica para tratar as variáveis numéricas com grande amplitude, uma vez que:
- já utilizaremos um conjunto grande de técnicas, o que afetará praticamente todos os atributos. Manter a forma de distribuição, como faz a padronização seria um processo um pouco menos disruptivo.
- as features que serão tratadas com a técnica de PCA também passarão por padronização. Aplicar também a PADRONIZAÇÃO no lugar da normalização nesse
conjunto de atributos garante uma maior consistência nos tratamentos.

#### Aplicando as demais técnicas de pré-processamento

Uma vez definidas as técnicas a serem utilizadas e os conjuntos que passarão por cada uma delas, vamos construir os pipelines para o modelo de machine learning.

```python
# Pipeline do frequency encoding
pipeline_freq = Pipeline([
    ("freq_encoder", FunctionTransformer(apply_frequency_encoding, validate=False))
    ])

# Pipeline do OneHotEncoder
pipeline_onehot = Pipeline ([
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

# Pipeline da Padronização
pipeline_norm = Pipeline([
    ('scaler', StandardScaler())
])

# Pipeline da PCA
pipeline_pca = Pipeline([
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=2))
])
```

```python
# Aplica ColumnTransformer unindo todos os pipelines
preprocessador = ColumnTransformer(transformers=[
    ('onehot', pipeline_onehot,col_hot),
    ('freq', pipeline_freq, col_frequency),
    ('norm', pipeline_norm, col_normalizacao),
    ('pca', pipeline_pca, col_pca)
], remainder = 'passthrough')
```

```python
# Aplica fit e transform no conjunto de treino
X_train_processado = preprocessador.fit_transform(X_train)
```

```python
# Aplica apenas transform no conjunto de teste
X_test_processado = preprocessador.transform(X_test)
```
